```python
import numpy as np

class LinkedList():
    def __init__(self, head = None):
        self.head = head
    
    
    def add_node_to_head(self, cargo):
        node = Node(cargo)
        node.next = self.head
        self.head = node
        
    
    def add_node_to_tail(self, cargo):
        node = Node(cargo)
        if self.head == None:
            self.head = node
        else:
            key_node = self.head
            while key_node.next != None:
                key_node = key_node.next
            key_node.next = node
        
        
    def peek_at_nth_node(self, n):
        length = self.show_length()
        if n > length: return None
        key_node = self.head
        counter = 0
        while key_node != None:
            if counter == n:
                nth_node = key_node
            key_node = key_node.next
            counter += 1
        return nth_node.cargo
        
         
    def delete_n_nodes_after_m_nodes(self, n, m):
        # like delete one node after 3 nodes until the end repetitively
        length = self.show_length()
        if length <= (m + n): return
        marker = self.head
        while marker != None:
            for i in range(m):
                if marker != None:
                    marker = marker.next
            for i in range(n):
                if marker != None:
                    removal_cargo = marker.cargo 
                    self.remove_item_with_cargo(removal_cargo)
                    marker = marker.next
             
        
      
    def show_length(self):
        if self.head == None: return 0
        key_node = self.head
        index_counter = 0
        while key_node.next != None:
            key_node = key_node.next
            index_counter += 1
        index_counter += 1
        print('the length of the linked list is:{}'.format(index_counter))
        return int(index_counter)
    
    
    def show_length_second_version(self):
        key_node = self.head
        index_counter = 0
        while key_node != None:
            index_counter += 1
            key_node = key_node.next
        return int(index_counter)
    
    
    def pop_first_item(self): 
        if self.head == None: return 
        popped_head = self.head
        self.head = self.head.next
        return popped_head.cargo
        
        
    def pop_last_item(self):
        if self.head.next == None:
            self.head = None
            return
        key_node = self.head
        next_node = key_node.next
        while next_node.next != None:
            next_node = next_node.next
            key_node = key_node.next 
        popped_item = next_node.cargo 
        key_node.next = None
        return popped_item

    
    def peek_at_tail(self):
        if self.head == None: return None
        key_node = self.head
        while key_node.next != None:
            key_node = key_node.next
        return key_node.cargo
                 

    def push_from_head(self, cargo):
        self.add_node_to_head(cargo)

        
    def push_from_tail(self, cargo):
        self.add_node_to_tail(cargo)
                
 
    def print_nodes(self):
        print('[')
        key_node = self.head
        while key_node != None:
            print(key_node.cargo)
            key_node = key_node.next
        print(']')
        
        
    def print_backwards(self):
        # create a new empty linked_list that takes self.head.next then recursively call the nodes cargo
        if self.head == None: return None
        new_list = LinkedList()
        new_list.head = self.head.next
        new_list.print_backwards()
        print(self.head.cargo)
        
        
    
    def insert_item_with_index(self, cargo, index): 
        new_node = Node(cargo)
        length = self.show_length()
        if index > length: 
            print('index is exceeding the length')
            return None
        if index == 0:
            if self.head == None:
                self.head = new_node
            else:
                new_node.next = self.head
                self.head = new_node
            return
        counter = 1
        pointer_node = self.head
        insertion_node = pointer_node.next
        while insertion_node != None:
            if counter == index:
                new_node.next = pointer_node.next
                pointer_node.next = new_node
                return
            insertion_node = insertion_node.next
            pointer_node = pointer_node.next
            counter += 1
            
        
    def remove_item_with_index(self, index): 
        length = self.show_length()
        if index >= length:
            print('index requested is exceeding the lenght')
            return
        if self.head == None: return
        if index == 0:
            self.head = self.head.next
            return
    
        pointer_node = self.head
        node_to_be_removed = pointer_node.next
        counter = 1
        assert(length > 1)
        while node_to_be_removed.next != None:
            if counter == index:
                print('{} is removed with index: {}'.format(node_to_be_removed.cargo, index))
                pointer_node.next = node_to_be_removed.next
                return
            node_to_be_removed = node_to_be_removed.next
            pointer_node = pointer_node.next
            counter += 1
        
        
    def remove_item_with_cargo(self, cargo): 
        if self.head == None: return 
        if self.head.cargo == cargo: 
            self.head = self.head.next
            return
        pointer_node = self.head
        node_to_be_removed = pointer_node.next
        while node_to_be_removed != None:
            if node_to_be_removed.cargo == cargo:
                print('this item is removed: {}'.format(node_to_be_removed.cargo))
                pointer_node.next = node_to_be_removed.next    
                return
            node_to_be_removed = node_to_be_removed.next
            pointer_node = pointer_node.next       
        print('the item is not found')
        return None
     
        
    def remove_duplicates(self): # remove duplicates next to each other
        length = self.show_length()
        if length < 2: return
        marker_node = self.head 
        while marker_node != None and marker_node.next != None:
            target_cargo = marker_node.cargo
            iter_node = marker_node
            while iter_node.next != None:
                if iter_node.next.cargo == target_cargo:
                    print('a duplicate item is removed with cargo: {}'.format(iter_node.next.cargo))
                    iter_node.next = iter_node.next.next
                else:
                    iter_node = iter_node.next
            marker_node = marker_node.next
            
            
    def find_the_frequency_of_items(self): 
        if self.head == None: return 
        key_node = self.head
        dictionary = {}
        while key_node != None:
            if key_node.cargo in dictionary.keys():
                dictionary[key_node.cargo] += 1
            else:
                dictionary[key_node.cargo] = 1
            key_node = key_node.next
        print("Frequency:")
        for k,v in dictionary.items():
            print(k,v)
        return dictionary
    
    
    # Just for challenge.. 
    def find_middle_and_split_list_in_two_and_delete_the_middlenode_if_odd_length(self):
        length = self.show_length()
        item_counter = 0
        key_node = self.head 
        if length % 2 == 0:
            print('the lenght is an even number, the left side and right side will have same length')
            left_side = LinkedList()
            right_side = LinkedList()
            mid_point = length // 2
            while key_node.next != None and item_counter < mid_point:
                left_side.add_node_to_tail(key_node.cargo)
                item_counter += 1
                key_node = key_node.next
            while key_node.next != None:
                right_side.add_node_to_tail(key_node.cargo)
                key_node = key_node.next
            right_side.add_node_to_tail(key_node.cargo)
            left_side.print_nodes()
            right_side.print_nodes()
            return (left_side , right_side)
            
        if length % 2 == 1:
            print('the lenght is an odd number, the exact middle item is skipped and split is done')
            left_side = LinkedList()
            right_side = LinkedList()
            mid_point = length // 2
            while key_node.next != None and item_counter < mid_point:
                left_side.add_node_to_tail(key_node.cargo)
                item_counter += 1
                key_node = key_node.next
            print('skipped item is:{}'.format(key_node.next.cargo))
            key_node = key_node.next #  skipping the middle item which is key_node.next
            while key_node.next != None:
                right_side.add_node_to_tail(key_node.cargo)
                key_node = key_node.next
            right_side.add_node_to_tail(key_node.cargo)
            left_side.print_nodes()
            right_side.print_nodes()
            return (left_side , right_side)
    
    
    def split_from_middle(self):
        length = self.show_length()
        item_counter = 0
        key_node = self.head 
        left_side = LinkedList()
        right_side = LinkedList()
        mid_point = length // 2
        while key_node != None and item_counter < mid_point:
            left_side.add_node_to_tail(key_node.cargo)
            item_counter += 1
            key_node = key_node.next
        if length % 2 == 1:
            print('the lenght is an odd number, the exact middle item is skipped and split is done')
            print('skipped item is:{}'.format(key_node.next.cargo))
            key_node = key_node.next #  skipping the middle item which is key_node.next
        else:
            print('the lenght is an even number, the left side and right side will have same length')
        while key_node != None:
            right_side.add_node_to_tail(key_node.cargo)
            key_node = key_node.next
        left_side.print_nodes()
        right_side.print_nodes()
        return (left_side , right_side)
    
    
    def split_middle_second_version(self):
        length = self.show_length()
        item_counter = 0
        key_node = self.head 
        left_side = LinkedList()
        right_side = LinkedList()
        mid_point = length // 2
        is_odd = length % 2 == 1
        curr_side = left_side
        while key_node != None:
            if item_counter == mid_point:
                curr_side = right_side
            skip_node = is_odd and item_counter == mid_point
            if not skip_node:
                curr_side.add_node_to_tail(key_node.cargo)
            item_counter += 1
            key_node = key_node.next
        left_side.print_nodes()
        right_side.print_nodes()
        return (left_side , right_side)


    def reverse_a_linked_list(self):
        reversed_list = LinkedList()
        key_node = self.head
        while key_node.next != None:
            reversed_list.add_node_to_head(key_node.cargo)
            key_node = key_node.next
        reversed_list.add_node_to_head(key_node.cargo)
        print('the list is reversed')
        reversed_list.print_nodes()
        return reversed_list
        
    
 
    def is_palindrome(self): 
        print('find the middle point, split, reverse the SECOND part of the list, check if both are the same')
        left, right = self.find_middle_and_split_list_in_two_and_delete_the_middlenode_if_odd_length()
        right = right.reverse_a_linked_list()
        left_key_node = left.head
        right_key_node = right.head
        while left_key_node.next != None and right_key_node.next != None:
            if left_key_node.cargo == right_key_node.cargo:
                left_key_node = left_key_node.next
                right_key_node = right_key_node.next
            else:
                print('the linked list is not palindrome')
                return False
        print('the linked list is palindrome')
        return True

        
    def is_palindrome_second_version(self):
        # make a new list, reverse it and compare with the original
        new_list = self.reverse_a_linked_list()
        key_node_original = self.head
        key_node_reversed = new_list.head
        while key_node_original != None:
            if key_node_original.cargo != key_node_reversed.cargo:
                return False
            key_node_original = key_node_original.next
            key_node_reversed = key_node_reversed.next
        return True
            
  
        
    def reverse_every_k_nodes(self, k):
        length = self.show_length()
        if k <= 1: return
        # create a new linked list
        n_repetions = length // k
        n_remaining_items = length % k
        new_list = LinkedList()
        key_node = self.head
        for i in range(n_repetions):
            small_list = LinkedList()
            for i in range(k):
                small_list.add_node_to_head(key_node.cargo)
                key_node = key_node.next
            small_list_key_node = small_list.head
            for i in range(k):  
                new_list.add_node_to_tail(small_list_key_node.cargo)
                small_list_key_node = small_list_key_node.next
        small_list = LinkedList()
        for i in range(n_remaining_items):
            small_list.add_node_to_head(key_node.cargo)
            key_node = key_node.next
        small_list_key_node = small_list.head
        for i in range(n_remaining_items):
            new_list.add_node_to_tail(small_list_key_node.cargo)
            small_list_key_node = small_list_key_node.next
        new_list.print_nodes()  
        
        
    def reverse_every_k_nodes_second_version(self, k):
        length = self.show_length()
        if k <= 1: return
        # create a new linked list
        n_repetitions = length // k
        n_remaining_items = length % k
        new_list = LinkedList()
        key_node = self.head
        n_to_reverse = k
        for i in range(n_repetitions + 1):
            small_list = LinkedList()
            if i == n_repetitions:
                n_to_reverse = n_remaining_items
            for i in range(n_to_reverse):
                small_list.add_node_to_head(key_node.cargo)
                key_node = key_node.next
            small_list_key_node = small_list.head
            for i in range(n_to_reverse):  
                new_list.add_node_to_tail(small_list_key_node.cargo)
                small_list_key_node = small_list_key_node.next
        new_list.print_nodes() 
        
        
    def create_bunch_of_test_nodes(self):
        list_of_items = np.arange(5)
        list_of_items = list(list_of_items)
        for i in list_of_items:
            self.add_node_to_tail(i)
            self.add_node_to_head(i)
       
    
    
class Node():
    def __init__(self, cargo, next = None):
        self.cargo = cargo
        self.next = next
        


new_list = LinkedList()
new_list.create_bunch_of_test_nodes()
new_list.print_nodes()
```

```python
# convert linkedList into a queue by class inheritance

class Queue(LinkedList): # FIFO
    # this class will push the item tail of the linked list, pop from head and will take a peek at the FIRST item
    def __init__(self, head = None):
        LinkedList.__init__(self, head)
        
    
    def push(self,cargo):
        # this method will push the product from tail
        self.add_node_to_tail(cargo)
    
    
    def peek(self): # this is returning the first item since it is a queue
        return self.peek_at_nth_node(0)
    
    
    def pop(self):
        # this method will pop the product from head
        return self.pop_first_item()
    
    
    def print_the_nodes(self):
        self.print_nodes()

    

queue_linked_list = Queue()
queue_linked_list.push(1)
queue_linked_list.push(2)
queue_linked_list.push(3)
#queue_linked_list.pop()
queue_linked_list.peek() 
queue_linked_list.print_the_nodes() 
```

```python
# import linkedlist properties into a stack class 

class Stack(): # LIFO
    def __init__(self):
        self.linkedList = LinkedList()
        
    
    def push(self, value):
        self.linkedList.push_from_head(value)
        
        
    def pop(self):
        # Stack is LIFO so we pop at the same end we push
        return self.linkedList.pop_first_item()
    
    
    def peek(self):
        # we take a peek at the head since we added and popped from head. Last item pushed is also returned 
        return self.linkedList.peek_at_nth_node(0) 
        
    def print(self):
        self.linkedList.print_nodes()
     
    
my_stack = Stack() # empty
my_stack.push(3) # 3
my_stack.push(5) # 3 5
my_stack.push(7) # 3 5 7
my_stack.peek()  # 7
#my_stack.print()

```
