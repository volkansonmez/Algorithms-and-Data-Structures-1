```python
import numpy as np

class BST():
    def __init__(self, root = None):
        self.root = root
       
    
    def add_node(self, value): # add node to a binary tree
        if self.root == None:
            self.root = Node(value)
        else:
            key_node = self.root
            self._add_node(self.root, value)
        
    
    def _add_node(self, key_node, value): # recursive function to place the nodes in place
        if key_node == None: return
        if value < key_node.cargo: # go left
            if key_node.leftchild == None:
                key_node.leftchild = Node(value)
                key_node.leftchild.parent = key_node
            else:
                self._add_node(key_node.leftchild, value)
        elif value > key_node.cargo:
            if key_node.rightchild == None:
                key_node.rightchild = Node(value)
                key_node.rightchild.parent = key_node
            else:
                self._add_node(key_node.rightchild, value)
        else: # if the value exists, do not add
            return
    
    
    def add_bunch_of_nodes_to_the_tree(self): # add random nodes to a binary tree
        nodes = np.arange(25)
        shuffled_nodes = np.random.permutation(nodes)
        for i in shuffled_nodes: 
            self.add_node(i)
        
        
    def create_a_V_tree(self): # create random numbers and add them to a V shape tree, one root, two leaves only
        random_numbers = np.arange(25)
        shuffled_nodes = np.random.permutation(random_numbers)
        for i in shuffled_nodes:
            self.create_a_V_shape_tree(i)
    
    
    def create_a_V_shape_tree(self, value): # create a V shape tree
        if self.root == None:
            self.root = Node(value)
        else:
            key_node = self.root
            self._create_a_V_shape_tree(key_node, value)

    
    def _create_a_V_shape_tree(self, key_node, value): # add nodes recursively to a V shape tree
        if key_node == None: return
        # split rules:
        if value < key_node.cargo: # go left on bst
            if key_node.leftchild == None: # if leftchild is None
                key_node.leftchild = Node(value)
                key_node.leftchild.parent = key_node
            elif value > key_node.leftchild.cargo: # squeeze the new node in between key_node and its leftchild
                new_node = Node(value)
                key_node.leftchild.parent = new_node
                new_node.leftchild = key_node.leftchild
                key_node.leftchild = new_node
                new_node.parent = key_node
            else: # if value is smaller than the leftchild's cargo
                key_node = key_node.leftchild
                self._create_a_V_shape_tree(key_node, value)
        elif value > key_node.cargo: # go right on bst
            if key_node.rightchild == None:
                key_node.rightchild = Node(value)
                key_node.rightchild.parent = key_node
            elif value < key_node.rightchild.cargo: # squeeze the new node in between key_node and its rightchild
                new_node = Node(value)
                key_node.rightchild.parent = new_node
                new_node.rightchild = key_node.rightchild
                key_node.rightchild = new_node
                new_node.parent = key_node
            else: # if the value is bigger than the rightchild's cargo
                key_node = key_node.rightchild
                self._create_a_V_shape_tree(key_node, value)
        else: # if the value exists already
            # print('The value already exists')
            return
            
        
        
    def find_node(self, value): # return the node object
        if self.root == None: return None
        if self.root.cargo == value:
            return self.root
        else:
            return self._find_node(self.root, value)
            
            
    def _find_node(self, key_node, value): # find and return the node recursively (not the cargo)
        if key_node == None: return None
        if value == key_node.cargo: return key_node
        if value < key_node.cargo:
            return self._find_node(key_node.leftchild, value)
        else:
            return self._find_node(key_node.rightchild, value)
        
    
            
    def print_in_order(self): # from left bottom leaf to the right bottom leaf, print all nodes 
        if self.root == None: return
        else:
            self._print_in_order(self.root)
            
    
    def _print_in_order(self, key_node): # recursive function to print the nodes in order
        if key_node != None:
            self._print_in_order(key_node.leftchild)
            print(key_node.cargo)
            self._print_in_order(key_node.rightchild)
    
    
    def list_all_nodes_in_order(self): # similar to print_in_order method, return a list of cargos in order
        all_nodes = []
        if self.root == None: return
        else:
            return self._list_all_nodes_in_order(all_nodes, self.root)
        
        
    def _list_all_nodes_in_order(self, all_nodes, key_node): # recursive function to return the list of cargos
        if key_node != None:
            self._list_all_nodes_in_order(all_nodes, key_node.leftchild)
            all_nodes.append(key_node.cargo)
            self._list_all_nodes_in_order(all_nodes, key_node.rightchild)
        return all_nodes
        
        
    def print_left_side(self): # print left side of the tree
        if self.root == None: return 
        else:
            self._print_left_side(self.root)
            
            
    def _print_left_side(self, key_node): # recursively percolate down and print the leftchilds
        if key_node == None: return
        self._print_left_side(key_node.leftchild)
        print(key_node.cargo , end = " ")
        
    
    def print_leaf_nodes_in_order_by_stacking(self): # alternative methodf to print the nodes in order
        all_nodes = [] # Stack to store all the nodes # of tree  
        leaf_nodes = [] # Stack to store all the # leaf nodes  
        
        all_nodes.append(self.root)  # Push the root node  

        while len(all_nodes) > 0:  
            key_node = all_nodes.pop()   
            if key_node.leftchild: 
                all_nodes.append(key_node.leftchild) 
            if key_node.rightchild:  
                all_nodes.append(key_node.rightchild)  
                
            elif key_node.leftchild == None and key_node.rightchild == None:
                leaf_nodes.append(key_node.cargo)  

        return leaf_nodes

            
    def print_leaf_nodes_in_order(self): # dfs to print nodes in order from left leaf to the right leaf
        if self.root == None: return
        else:
            self._print_leaf_nodes_in_order(self.root)
            
    
    def _print_leaf_nodes_in_order(self, key_node): # recursively go left, print the leaf node, go right
        if key_node != None:
            self._print_leaf_nodes_in_order(key_node.leftchild)
            if key_node.leftchild == None and key_node.rightchild == None:
                print(key_node.cargo)
            self._print_leaf_nodes_in_order(key_node.rightchild)
       
    
    def find_height_of_bst(self): # finding the height of the BST
        if self.root == None: return 0
        else:
            leftheight, rightheight = 0, 0
            key_node = self.root
            return self._find_height_of_bst(key_node, leftheight, rightheight)
    
    
    def _find_height_of_bst(self, key_node, leftheight, rightheight): # recursive function to return the height of BST
        if key_node == None: 
            return max(leftheight, rightheight)
        else:
            return self._find_height_of_bst(key_node.leftchild , leftheight + 1, rightheight)
            return self._find_height_of_bst(key_node.rightchild, leftheight, rightheight + 1)  
    
    
    def print_height(self): # function to print the height
        height_of_tree = self.find_height_of_bst()
        print('the height of the tree is: {}'.format(height_of_tree))
    
    
    def height_balance(self): # height method to find if the tree is height balanced 
        if self.root == None: return 0
        left_branch = self.root.leftchild
        right_branch = self.root.rightchild
        return self.find_height(left_branch), self.find_height(right_branch)
        
    
    def find_height(self, branch): # finding the height of the BST branch
        if branch == None: return 0
        else:
            leftheight, rightheight = 0, 0
            key_node = branch
            return self._find_height_of_bst(key_node, leftheight, rightheight)
    
    
    def _find_height_of_bst(self, key_node, leftheight, rightheight): # recursive function to return the height of BST
        if key_node == None: 
            return max(leftheight, rightheight)
        else:
            return self._find_height_of_bst(key_node.leftchild , leftheight + 1, rightheight)
            return self._find_height_of_bst(key_node.rightchild, leftheight, rightheight + 1)  
    
    
    def print_height_balance(self): # print the height balance
        left_and_right_heights = self.height_balance()
        print('the height of left and right branches: {}'.format(left_and_right_heights))
        
    
    def delete(self, value): # delete a node
        node_to_be_deleted = self.find_node(value)
        if node_to_be_deleted == None: return None
        else:
            return self._delete(node_to_be_deleted, value)
    
    
    def _delete(self, key_node, value): # function to find the number of children and delete the node accordingly
    
        def find_children(key_node): # nested function to find the number of children
            n = 0
            if key_node.leftchild != None:
                n +=1
            if key_node.rightchild != None:
                n +=1
            return n
        
        number_of_children = find_children(key_node)
        
        
        def find_min_in_tree(key_node): # find the min value in the tree
            while key_node.leftchild != None:
                key_node = key_node.leftchild
            return key_node
        
        if key_node.rightchild != None:
            successor = find_min_in_tree(key_node.rightchild)
        
        
        # delete nodes for key_node having no children
        if number_of_children == 0:
            # it is a leaf node
            if key_node.parent.leftchild == key_node:
                key_node.parent.leftchild = None
            else:
                key_node.parent.rightchild = None
                
                
        if number_of_children == 1:
            # there are 4 cases:
            # the node is parent's leftchild
                # the node has leftchild
            if key_node.parent.leftchild == key_node and key_node.leftchild != None:
                key_node.parent.leftchild = key_node.leftchild
                key_node.leftchild.parent = key_node.parent
                # the node has rightchild
            elif key_node.parent.leftchild == key_node and key_node.rightchild != None:
                key_node.parent.leftchild = key_node.rightchild
                key_node.rightchild.parent = key_node.parent
            # the node is parent's rightchild
                # the node has leftchild
            elif key_node.parent.rightchild == key_node and key_node.leftchild != None:
                key_node.parent.rightchild = key_node.leftchild
                key_node.leftchild.parent = key_node.parent
            else:
                # the node has rightchild
                key_node.parent.rightchild = key_node.rightchild
                key_node.rightchild.parent = key_node.parent
                
                
        if number_of_children == 2:
            successor = find_min_in_tree(key_node.rightchild)
            key_node.cargo = successor.cargo
            successor = None
        
        
    def is_valid(self): # check if the tree not only has validity for the value order between parent child nodes 
        if self.root == None: return True
        else:
            return self._is_valid(self.root, -np.inf, np.inf)

        
    def _is_valid(self, key_node, min, max): # recursive function to return the cargo values if between min and max
        if key_node.cargo <= min or key_node.cargo >= max: return False
        left_ok = True
        right_ok = True
        if key_node != None and key_node.leftchild != None:
            left_ok = self._is_valid(key_node.leftchild, min, key_node.cargo)
        if key_node != None and key_node.rightchild != None:
            right_ok = self._is_valid(key_node.rightchild, key_node.cargo, max)
        return left_ok and right_ok
        
        
    def print_validity(self):  # function to print the validity of the bst
        is_valid_bst = self.is_valid()
        if is_valid_bst:
            print('the bst is valid')
        else:
            print('the tree is not a bst')


    def print_bfs(self):  # it is to print the nodes in vertical order, BFS
        print('Starting the breadth first search printing')
        if self.root == None: return None
        key_node = self.root
        self._print_bfs(self.root)
           
            
    def _print_bfs(self, key_node, todo = None): # recursive function using stack to print each level of nodes
        if todo == None: todo = []
        if key_node == None: return 
        print(key_node.cargo)
        if key_node.leftchild != None:
            todo.append(key_node.leftchild)
        if key_node.rightchild != None:
            todo.append(key_node.rightchild)
        if len(todo) > 0:
            next_node = todo.pop(0)
            self._print_bfs(next_node , todo)
            
            
    def dfs(self): # print in order with depth first search 
        if self.root == None: return
        else:
            key_node = self.root
            self.print_dfs(key_node)
            
            
    def print_dfs(self, key_node): # recursive function to print the node cargos in order with DFS left to right
        if key_node == None: return None
        if key_node.leftchild != None: self.print_dfs(key_node.leftchild)
        print(key_node.cargo)
        if key_node.rightchild != None: self.print_dfs(key_node.rightchild)
        
    
    def print_pre_order(self): # print the node cargos in pre-order 
        if self.root == None: return None
        else:
            self._print_pre_order(self.root)
        
    
    def _print_pre_order(self, key_node, todo = None): # recursive function to print the nodes in pre-order
        if todo == None: todo = []
        if key_node == None: return 
        print(key_node.cargo)
        if key_node.rightchild != None:
            todo.insert(0, key_node.rightchild)
        if key_node.leftchild != None:
            todo.insert(0, key_node.leftchild)
        if len(todo) > 0:
            popped_item = todo.pop(0)
            self._print_pre_order(popped_item, todo)
            
    
    def zig_zag_printing_top_to_bottom(self): # spiral order printing using two stacks 
        if self.root == None: return
        even_stack = [] # stack the nodes in levels that are in even numbers
        odd_stack = [] # stack the nodes in levels that are in odd numbers
        print_nodes = [] # append the items' cargos in zigzag order
        even_stack.append(self.root)
        while len(even_stack) > 0 or len(odd_stack) > 0:
            
            while len(even_stack) > 0:
                tmp = even_stack.pop()
                print_nodes.append(tmp.cargo)
                if tmp.rightchild != None:
                    odd_stack.append(tmp.rightchild)
                if tmp.leftchild != None:
                    odd_stack.append(tmp.leftchild)
    
    
            while len(odd_stack) > 0:
                tmp = odd_stack.pop()
                print_nodes.append(tmp.cargo)
                if tmp.leftchild != None:
                    even_stack.append(tmp.leftchild)
                if tmp.rightchild != None:
                    even_stack.append(tmp.rightchild)
        
        return print_nodes
    
    
    def lowest_common_ancestor(self, node1, node2): # takes two cargos and prints the lca 
        if self.root == None: return
        node1_confirm = self.find_node(node1)
        if node1_confirm == None: return
        node2_confirm = self.find_node(node2)
        if node2_confirm == None: return
        key_node = self.root
        print('nodes are in the tree')
        return self._lowest_common_ancestor(key_node, node1, node2)
    
    
    def _lowest_common_ancestor(self, key_node, node1, node2): # recursive function to return the lcs of two nodes
        if key_node == None: return
        if node1 < key_node.cargo and node2 < key_node.cargo:
            key_node = key_node.leftchild
            return self._lowest_common_ancestor(key_node, node1, node2)
        elif node1 > key_node.cargo and node2 > key_node.cargo:
            key_node = key_node.rightchild
            return self._lowest_common_ancestor(key_node, node1, node2)
        else:
            return key_node , key_node.cargo
        
    
    def maximum_path_sum(self): # function to find the maximum path sum
        if self.root == None: return
        max_value = -np.inf
        return self._maximum_path_sum(self.root, max_value)
        
        
    def _maximum_path_sum(self, key_node, max_value): # recursive function to search and return the max path sum
        if key_node == None: return 0
        left = self._maximum_path_sum(key_node.leftchild, max_value)
        right = self._maximum_path_sum(key_node.rightchild, max_value)
        max_value = max(max_value, key_node.cargo + left + right)
        return max(left, right) + self.root.cargo

    
    
    
class Node():
    def __init__(self, cargo, parent = None, leftchild = None, rightchild = None):
        self.cargo = cargo
        self.parent = parent
        self.leftchild = leftchild
        self.rightchild = rightchild
        

        
        
      
test_bst = BST()
#test_bst.add_node(30)
test_bst.add_bunch_of_nodes_to_the_tree()
#test_bst.print_height()
#test_bst.print_in_order()
#test_bst.print_validity()
#test_bst.find_node(24)   
#test_bst.delete(27)
#test_bst.print_bfs()
#test_bst.print_pre_order()
#test_bst.print_dfs(test_bst.root)
#test_bst.print_left_side()
#test_bst.print_leaf_nodes_in_order_by_stacking()     
test_bst.print_leaf_nodes_in_order()
#test_bst.list_all_nodes_in_order()
#test_bst.zig_zag_printing_top_to_bottom()
#test_bst.print_height_balance()
#test_bst.maximum_path_sum()
#test_bst2 = BST() # this is going to be a V shaped structure, having 2 leaf nodes only.
#test_bst2.create_a_V_tree()
#test_bst2.print_leaf_nodes_in_order()
```
